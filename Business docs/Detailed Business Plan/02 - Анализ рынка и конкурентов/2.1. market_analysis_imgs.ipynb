```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#  
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))

#    (    )
regions = ['. ', '', '', ' ']
current_share = np.array([35, 30, 25, 10])
growth_rate = np.array([12, 14, 18, 10])  # CAGR  

#  subplot layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# 1.    
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']
wedges, texts, autotexts = ax1.pie(
    current_share, 
    labels=regions,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'w', 'linewidth': 2},
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)

#   
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

ax1.set_title('    , 2023', fontsize=14, fontweight='bold')

# 2.   CAGR  
bars = ax2.bar(
    regions, 
    growth_rate, 
    color=colors,
    width=0.6,
    edgecolor='black',
    linewidth=1.5
)

#    
for bar in bars:
    height = bar.get_height()
    ax2.text(
        bar.get_x() + bar.get_width()/2., 
        height + 0.5,
        f'{height:.1f}%',
        ha='center', 
        va='bottom', 
        fontsize=12, 
        fontweight='bold'
    )

ax2.set_title('   (CAGR)  , 2023-2028', fontsize=14, fontweight='bold')
ax2.set_ylabel('CAGR, %', fontsize=14)
ax2.tick_params(axis='x', rotation=30)
ax2.set_ylim(0, 25)

#  
plt.figtext(0.5, 0.01, ':     Grand View Research, MarketsandMarkets, Mordor Intelligence', 
           ha='center', fontsize=10, style='italic')

plt.tight_layout(pad=3)
plt.savefig('regional_market_mt.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#  
sns.set(style="whitegrid")

#   
industries = [
    '  ', 
    '  ', 
    ' ', 
    ' ', 
    ' ', 
    '  ', 
    ''
]

market_share_2023 = np.array([28, 17, 12, 11, 9, 8, 15])  #  
growth_rate = np.array([16, 14, 13, 11, 15, 17, 10])  # CAGR  

#  dataframe  
df = pd.DataFrame({
    '': industries,
    ' , %': market_share_2023,
    'CAGR, %': growth_rate
})

#    
df = df.sort_values(' , %', ascending=False)

#  subplot layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))

#   
colors = sns.color_palette("viridis", len(industries))

# 1.     
bars1 = ax1.barh(df[''], df[' , %'], color=colors, height=0.6, edgecolor='black', linewidth=1)

#     
for i, bar in enumerate(bars1):
    width = bar.get_width()
    ax1.text(width + 0.5, bar.get_y() + bar.get_height()/2, 
            f'{width:.1f}%', ha='left', va='center', fontsize=12, fontweight='bold')

ax1.set_title('     , 2023', fontsize=16, fontweight='bold')
ax1.set_xlabel(' , %', fontsize=14)
ax1.set_xlim(0, 35)  #    X    

# 2.  :   vs CAGR
bubble_sizes = df[' , %'] * 50  #    

scatter = ax2.scatter(
    df['CAGR, %'], 
    range(len(df[''])), 
    s=bubble_sizes, 
    c=colors,
    alpha=0.7,
    edgecolor='black',
    linewidth=1.5
)

#    
for i, row in df.iterrows():
    ax2.text(row['CAGR, %'] + 0.3, df[''].tolist().index(row['']), 
             f'{row["CAGR, %"]:.1f}%', va='center', fontsize=11, fontweight='bold')

ax2.set_yticks(range(len(df[''])))
ax2.set_yticklabels(df[''])
ax2.set_title('    (CAGR, 2023-2028)', fontsize=16, fontweight='bold')
ax2.set_xlabel('CAGR, %', fontsize=14)
ax2.set_xlim(9, 20)  #    X

#     
sizes = [10, 20, 30]
labels = ['10%', '20%', '30%']
legend_bubbles = []
for size in sizes:
    legend_bubbles.append(plt.scatter([], [], s=size*50, c='gray', alpha=0.7, edgecolor='black', linewidth=1.5))

ax2.legend(legend_bubbles, labels, title=" ", loc="lower right", title_fontsize=12)

#   
plt.suptitle('       ', fontsize=18, fontweight='bold', y=0.98)

#  
plt.figtext(0.5, 0.01, ':     Grand View Research, MarketsandMarkets, Mordor Intelligence', 
           ha='center', fontsize=10, style='italic')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('industry_market_mt.png', dpi=300, bbox_inches='tight')
plt.show()
```
