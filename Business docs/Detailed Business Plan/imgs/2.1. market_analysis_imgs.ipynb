```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Настройка стиля
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# Данные о росте рынка (в млн долларов США)
years = np.array([2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028])
market_size = np.array([650, 720, 800, 910, 1030, 1160, 1310, 1480, 1675, 1900])

# Реальные данные до 2023, прогнозные после
actual_years = years[:5]
actual_market = market_size[:5]
forecast_years = years[4:]
forecast_market = market_size[4:]

# Основной график
plt.plot(actual_years, actual_market, 'o-', color='#1f77b4', linewidth=3, markersize=10, label='Фактические данные')
plt.plot(forecast_years, forecast_market, 'o--', color='#ff7f0e', linewidth=3, markersize=10, label='Прогнозные данные')

# Добавление CAGR
cagr = ((market_size[-1] / market_size[0]) ** (1 / (len(years) - 1)) - 1) * 100
plt.text(2023.5, 1400, f'CAGR (2019-2028): {cagr:.1f}%', fontsize=12, fontweight='bold')

# Аннотации ключевых технологических событий
plt.annotate('Прорыв в NMT', xy=(2020, 720), xytext=(2018.5, 850),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
            fontsize=10, fontweight='bold')

plt.annotate('Интеграция с LLM', xy=(2023, 1030), xytext=(2021.5, 1200),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8),
            fontsize=10, fontweight='bold')

# Настройка осей и заголовка
plt.title('Динамика роста глобального рынка машинного перевода (2019-2028)', fontsize=16, fontweight='bold')
plt.xlabel('Год', fontsize=14)
plt.ylabel('Объём рынка (млн долларов США)', fontsize=14)
plt.xticks(years, fontsize=12)
plt.yticks(fontsize=12)

# Добавление сетки и легенды
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Диаграмма технологий (внутренняя)
plt.axes([0.58, 0.2, 0.3, 0.3])
technologies = ['Нейронный МТ', 'Статистический МТ', 'Гибридный МТ', 'На основе правил']
shares = [65, 20, 10, 5]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
plt.pie(shares, labels=technologies, autopct='%1.0f%%', colors=colors, 
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'})
plt.title('Доли технологий МТ в 2023 году', fontsize=12)

plt.tight_layout()
plt.savefig('market_growth_mt.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Настройка стиля
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))

# Данные по регионам (в процентах от общего рынка)
regions = ['Сев. Америка', 'Европа', 'АТР', 'Остальной мир']
current_share = np.array([35, 30, 25, 10])
growth_rate = np.array([12, 14, 18, 10])  # CAGR по регионам

# Создаем subplot layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# 1. Пирограмма текущей доли рынка
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']
wedges, texts, autotexts = ax1.pie(
    current_share, 
    labels=regions,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'w', 'linewidth': 2},
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)

# Изменение стиля текста
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

ax1.set_title('Региональная структура рынка машинного перевода, 2023', fontsize=14, fontweight='bold')

# 2. Столбчатая диаграмма CAGR по регионам
bars = ax2.bar(
    regions, 
    growth_rate, 
    color=colors,
    width=0.6,
    edgecolor='black',
    linewidth=1.5
)

# Добавление значений над столбцами
for bar in bars:
    height = bar.get_height()
    ax2.text(
        bar.get_x() + bar.get_width()/2., 
        height + 0.5,
        f'{height:.1f}%',
        ha='center', 
        va='bottom', 
        fontsize=12, 
        fontweight='bold'
    )

ax2.set_title('Среднегодовой темп роста (CAGR) по регионам, 2023-2028', fontsize=14, fontweight='bold')
ax2.set_ylabel('CAGR, %', fontsize=14)
ax2.tick_params(axis='x', rotation=30)
ax2.set_ylim(0, 25)

# Добавление примечаний
plt.figtext(0.5, 0.01, 'Источник: анализ на основе данных Grand View Research, MarketsandMarkets, Mordor Intelligence', 
           ha='center', fontsize=10, style='italic')

plt.tight_layout(pad=3)
plt.savefig('regional_market_mt.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Настройка стиля
sns.set(style="whitegrid")

# Данные по отраслям
industries = [
    'ИТ и телекоммуникации', 
    'Здравоохранение и фармацевтика', 
    'Автомобильная промышленность', 
    'Государственный сектор', 
    'Финансовые услуги', 
    'Медиа и развлечения', 
    'Прочие'
]

market_share_2023 = np.array([28, 17, 12, 11, 9, 8, 15])  # в процентах
growth_rate = np.array([16, 14, 13, 11, 15, 17, 10])  # CAGR по отраслям

# Создаем dataframe для удобства
df = pd.DataFrame({
    'Отрасль': industries,
    'Доля рынка, %': market_share_2023,
    'CAGR, %': growth_rate
})

# Сортировка по доле рынка
df = df.sort_values('Доля рынка, %', ascending=False)

# Создаем subplot layout
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))

# Цвета для отраслей
colors = sns.color_palette("viridis", len(industries))

# 1. Горизонтальная столбчатая диаграмма долей рынка
bars1 = ax1.barh(df['Отрасль'], df['Доля рынка, %'], color=colors, height=0.6, edgecolor='black', linewidth=1)

# Добавление значений справа от столбцов
for i, bar in enumerate(bars1):
    width = bar.get_width()
    ax1.text(width + 0.5, bar.get_y() + bar.get_height()/2, 
            f'{width:.1f}%', ha='left', va='center', fontsize=12, fontweight='bold')

ax1.set_title('Доля отраслей на рынке машинного перевода, 2023', fontsize=16, fontweight='bold')
ax1.set_xlabel('Доля рынка, %', fontsize=14)
ax1.set_xlim(0, 35)  # Установим предел оси X немного выше максимального значения

# 2. Пузырьковая диаграмма: размер рынка vs CAGR
bubble_sizes = df['Доля рынка, %'] * 50  # Масштабирование для лучшей визуализации

scatter = ax2.scatter(
    df['CAGR, %'], 
    range(len(df['Отрасль'])), 
    s=bubble_sizes, 
    c=colors,
    alpha=0.7,
    edgecolor='black',
    linewidth=1.5
)

# Добавление подписей к пузырькам
for i, row in df.iterrows():
    ax2.text(row['CAGR, %'] + 0.3, df['Отрасль'].tolist().index(row['Отрасль']), 
             f'{row["CAGR, %"]:.1f}%', va='center', fontsize=11, fontweight='bold')

ax2.set_yticks(range(len(df['Отрасль'])))
ax2.set_yticklabels(df['Отрасль'])
ax2.set_title('Темп роста по отраслям (CAGR, 2023-2028)', fontsize=16, fontweight='bold')
ax2.set_xlabel('CAGR, %', fontsize=14)
ax2.set_xlim(9, 20)  # Установим пределы оси X

# Добавление легенды для размера пузырьков
sizes = [10, 20, 30]
labels = ['10%', '20%', '30%']
legend_bubbles = []
for size in sizes:
    legend_bubbles.append(plt.scatter([], [], s=size*50, c='gray', alpha=0.7, edgecolor='black', linewidth=1.5))

ax2.legend(legend_bubbles, labels, title="Доля рынка", loc="lower right", title_fontsize=12)

# Заголовок всей фигуры
plt.suptitle('Отраслевая структура рынка машинного перевода и темпы роста', fontsize=18, fontweight='bold', y=0.98)

# Добавление примечаний
plt.figtext(0.5, 0.01, 'Источник: анализ на основе данных Grand View Research, MarketsandMarkets, Mordor Intelligence', 
           ha='center', fontsize=10, style='italic')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('industry_market_mt.png', dpi=300, bbox_inches='tight')
plt.show()
```
